# File 1

```python
import numpy as np
import matplotlib.pyplot as plt

# get data
seismic_data = np.genfromtxt("data/file1.txt").astype(np.float32)
seismic_data.shape
```

```python
# visualize for good measure
time = [i for i in range(seismic_data.shape[0])]

fig, axs = plt.subplots(1, 1, sharex=True)
fig.set_size_inches(18.5, 6)

plt.plot(time, seismic_data.T, 'b')
```

# Change Point 1

```python
# put arrays in dictionary for labels
seis_data = {}

seis_data["channel_0"] = seismic_data.T
```

```python
from importlib import reload
import CPD.warm

reload(CPD.warm)

# warm up for 1 second
fftwarm = CPD.warm.ARPWarm(seis_data, n0=5, chunks=1)
fftwarm.warmup()
```

```python
import CPD.CPD
import CPD.benchmark
reload(CPD.benchmark)

cpd_earth = CPD.benchmark.Online(fftwarm, k=1, M=2)

change_points, _ = cpd_earth.run()

cpt = max(change_points.values())

print(f"change point detected at t={cpt}")
```

```python
# visualize for good measure

fig, axs = plt.subplots(1, 1, sharex=True)
fig.set_size_inches(18.5, 6)

axs.plot(time, seismic_data.T, 'b')

axs.axvline(x=cpt, c='g', linewidth=5)
```

# Change Point 2

```python
# get data past detected change-point
seis_data2 = {}
seis_data2["channel_0"] = seis_data["channel_0"][cpt:]
```
